Now we'll graph our *lines of best fit.* This is going to show us the trend in how our plants grow! We can do this with code, however it is important that we know how this works.

To do this, we'll need to add up each of our data points in both the x and y axis. To demonstrate, we'll use some made up data.

```python
add = np.array([1, 5, 8, 10]) # this is the numbers we will add together
sum = np.sum(add) # this shows us the sum of all the numbers in our add array
print(add[0], " + ", add[1], " + ", add[2], " + ", add[3])
numberpoints = 4
average = sum/numberpoints
print(average)
```

We can finally see our data! By running the following code we can see what our line of best fit is! We can also see how much each plant grow every day!

```python
plt.title("Title") # this is the title of our graph
plt.xlabel("X Axis") # this is the title of our x axis
plt.ylabel("Y Axis") # this is the title our our y axis
plt.plot(x, y, color = "red")
plt.plot(x, t, color = "blue")
m, b = np.polyfit(x, y, 1)
c, d = np.polyfit(x, t, 1)
plt.plot(x, m*x + b, color = "yellow")
plt.plot(x, c*x + b, color = "green")
print("The first plant grows an average of ", m, "units per day") # change units to represent what you measured in!
print("The first plant grows an average of ", c, "units per day") # change units to represent what you measured in!
```
