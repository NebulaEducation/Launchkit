```python
import numpy as np #Imports numpy library
import matplotlib.pyplot as plt #Imports matplot library
```

```python
x = np.arange(0,10) #This is our x-axis. This is how many days we are measuring the plant's growth
y = np.array([0, 0, 1, 2, 4, 7, 8, 15, 16, 18]) #This is our y-axis. This is the plant's height in inches every day
```

```python
plt.title("Plant Growth") #This is the title of our graph
plt.xlabel("X Axis") #This is the title of the x-axis of our graph
plt.ylabel("Y Axis") #This is the title of the y-axis of our graph
plt.plot(x, y, color = "red") #This graphs our line
plt.show() #This shows our graph
```

```python
plt.title("Plant Growth")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.plot(x, y, color = "red")
m, b = np.polyfit(x, y, 1) #This creates a line of best fit
plt.plot(x, m*x +b) #This plots our line of best fit
```
